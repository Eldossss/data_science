```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()

import statsmodels.api as sm 
```

```python
DATA = pd.read_csv('Admittance.csv')
```

```python
DATA
```

```python
DATA['Admitted'] = DATA['Admitted'].map({'Yes':1,'No':0})
```

```python
DATA
```

```python
y = DATA['Admitted']
x1 = DATA['SAT']
```

```python
x = sm.add_constant(x1)
res_log = sm.Logit(y,x).fit() 
```

```python
res_log.summary()
```
