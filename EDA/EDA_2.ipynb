```python
DATA['review_rating'] = pd.to_numeric(DATA['review_rating'], errors='coerce') # эта строка кода если в столбце есть пробелы там тексты превращает их в NaN потом мы опять сможем удалить NaN
```

```python
DATA = DATA.dropna()
```

```python
DATA = DATA.reset_index(drop = True)
```

```python
DATA
```

```python
DATA['review_rating'] = DATA['review_rating'].astype(float)   # so we converrted object type to float type
```

```python
DATA.groupby('course_name')['review_rating'].mean().sort_values(ascending = False)
```

```python
DATA.groupby('course_name')['review_rating'].mean().sort_values(ascending = True)
```

```python
# we can notice that our the lowest grade course is "How to think Like a Data Scientist" and the highest rated course is "Introduction to Data Science"
```

```python
review_counts = DATA['course_name'].value_counts()
review_counts
plt.figure(figsize=(12, 6))
sns.barplot(x=review_counts.values, y=review_counts.index)
plt.title('Количество отзывов по каждому курсу')
plt.xlabel('Количество отзывов')
plt.ylabel('Название курса')
plt.tight_layout()
plt.show()
```
