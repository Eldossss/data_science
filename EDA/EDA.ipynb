```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df  = pd.read_csv('vehicles_us.csv')
```

```python
plt.style.use('seaborn-v0_8')
sns.set_palette("deep")
```

```python
# мы сперва даем настройки  для матплот либа , потом устанавливаем размер  графика на plt , потом мы строим гистогарамму 
plt.figure(figsize = (20,10))
sns.histplot(df['price'],bins = 50 , kde = True)
plt.title("Распределение цен на  автомобили")
plt.xlabel("цена")
plt.ylabel("количество")
plt.show()
```

```python
df.head(n = 10)
```

```python
new_price = df['price'] > 75000
plt.figure(figsize  = (10,6))
sns.histplot(df[new_price]['price'] , bins = 50, kde = True)
plt.title('Распределение цен на автмоили, новые данные')
plt.xlabel('Цены на автомобили')
plt.ylabel('Количество машин')
plt.show()
```
