```python
import numpy as np
import pandas as pd
import statsmodels.api as sm 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('real_estate_price_size.csv')
```

```python
data.describe()
```

```python
x1 = data['size']
y = data['price']
plt.scatter(x1,y)
plt.xlabel('SIZE', fontsize = 20)
plt.ylabel('PRICE',fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
x.shape
```

```python
XTX = np.dot(x.T,x)
XTX_inv = np.linalg.inv(XTX)
XTy = np.dot(x.T,y)
b = np.dot(XTX_inv,XTy)
print(b)
## this is hand-witten algorithm to find coeff b0, b1
## formula is B = (x^T*x)^inverse*(X^T*y) , where the x is array with 100 row and 2 column , 1 column is just bunch of ones . y is the real output values for each input
```

```python
plt.scatter(x1,y)
ythat = 1.019e+05 + x1*223.1787
line = plt.plot(x1,ythat,lw = 4 , c = 'red' ,label = 'regression line')
plt.xlabel('SIZE',fontsize = 30)
plt.ylabel('PRICE',fontsize = 30)
plt.show()
```
