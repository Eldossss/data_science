```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns

sns.set
```

```python
data = pd.read_csv('multiple_regression_model.csv')
```

```python
data 
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT','Rand 1,2,3']]
x1
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
# i guess rand 1,2,3 is insignificant predictor , because we can see than coefficient of that predictor is negative p-value > 0.05
# despite that let's compare table with only one predictor
```

```python
x1 = data['SAT']
y = data['GPA']
```

```python
x = sm.add_constant(x1)
lin_reg = sm.OLS(y,x).fit()
lin_reg.summary()
# btw OLS is method which is used in linear regression model least square sum , or min(SUM((y_i - y_pred)^2))
```

```python
# hand writen algorithm  
```

```python
XTX = np.dot(x.T,x)
XTX_inv = np.linalg.inv(XTX)
XTy = np.dot(x.T,y)
res = np.dot(XTX_inv,XTy)
print(res)
```

## NOTES

```python
# we can notice that when we used as predictor only SAT R^2 was 0.406, Adjusted R^2 was 0.399, but when we used as predictors unnecessary rand 1,2,3
# our R^2 increased to 0.407 , but Adjusted R^2 decreased to 0.392 , and Adjusted R^2 shows the exactly how model is good without unnecessary predictors
# if number of predictor will increase R^2 value also will ncrease , but Adjusted R^2 will decrease 
```
